```python
import pandas as pd
import matplotlib.pyplot as plt

# Results data
data = {
    "Model": ["XGBoost", "MLP"],
    "Original Acc": [0.868, 0.851],
    "Renamed Acc": [0.751, 0.751],
    "+Dict Map Acc": [0.868, 0.851],
    "+Embedding Acc": [0.806, 0.756]
}
df = pd.DataFrame(data)
df["Acc Drop (Renamed)"] = df["Original Acc"] - df["Renamed Acc"]
df["Acc Drop (+Embed)"] = df["Original Acc"] - df["+Embedding Acc"]

# Plot 1: Accuracy Comparison
df.set_index("Model")[["Original Acc", "Renamed Acc", "+Dict Map Acc", "+Embedding Acc"]].plot.bar()
plt.title("Model Accuracy Under Schema Shifts")
plt.ylabel("Accuracy")
plt.ylim(0.7, 0.9)
plt.xticks(rotation=0)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# Plot 2: Accuracy Drop
df.set_index("Model")[["Acc Drop (Renamed)", "Acc Drop (+Embed)"]].plot.bar(color=["salmon", "skyblue"])
plt.title("Accuracy Drop Due to Schema Shift")
plt.ylabel("Accuracy Drop")
plt.xticks(rotation=0)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Create DataFrame for tabular model results
data = {
    "Model": ["XGBoost", "MLP"],
    "Original Acc": [0.868, 0.851],
    "Renamed Acc": [0.751, 0.751],
    "+Dict Map Acc": [0.868, 0.851],
    "+Embedding Acc": [0.806, 0.756]
}
df = pd.DataFrame(data)

# Calculate accuracy drop
df["Acc Drop (Renamed)"] = df["Original Acc"] - df["Renamed Acc"]
df["Acc Drop (+Embed)"] = df["Original Acc"] - df["+Embedding Acc"]

# Style table
styled = df.style.set_caption("Results (Tabular Models)") \
    .format("{:.3f}", subset=df.columns[1:]) \
    .background_gradient(subset=["Original Acc", "Renamed Acc", "+Dict Map Acc", "+Embedding Acc"], cmap="Greens") \
    .background_gradient(subset=["Acc Drop (Renamed)", "Acc Drop (+Embed)"], cmap="OrRd")

styled
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Original schema sample
original_schema = pd.DataFrame({
    "Feature": ["workclass", "marital-status", "age", "capital-gain"],
    "Value": ["Private", "Divorced", "39", "0"]
})

# Shifted schema with renaming, merging, and composition
shifted_schema = pd.DataFrame({
    "Feature": ["col_workclass", "maritalstatus", "age_bracket", "cap_gain"],
    "Value": ["Private", "Formerly-Married", "Middle", "0"]
})

# Create the side-by-side tables
fig, axes = plt.subplots(1, 2, figsize=(12, 3))
axes[0].axis('off')
axes[1].axis('off')

# Display the original schema
axes[0].table(cellText=original_schema.values,
              colLabels=original_schema.columns,
              loc='center')
axes[0].set_title("Original Schema")

# Display the shifted schema
axes[1].table(cellText=shifted_schema.values,
              colLabels=shifted_schema.columns,
              loc='center')
axes[1].set_title("Shifted Schema")

plt.tight_layout()
plt.show()
```
